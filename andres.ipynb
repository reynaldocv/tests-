```python
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        
        for char in s: 
            if char in "([{":
                stack.append(char)
            elif char == ")":
                if stack and stack[-1] == "(":
                    stack.pop() 
                else: 
                    return False 
            elif char == "]":
                if stack and stack[-1] == "[":
                    stack.pop() 
                else: 
                    return False 
            elif char == "}":
                if stack and stack[-1] == "{":
                    stack.pop() 
                else: 
                    return False 
                        
        return len(stack) == 0
    
# Se va a analizar un string (str) que es una cadena de caracteres o sea es una palabra
# como por ejemplo "(([]))", "[]{}[][]{}", o "(abcdef)"

# El -> bool no es necesario. Las primeras version de python no usaban, ahora solo se coloca para saber 
# que valor va a ser retornado. Si tú especificas el valor -> bool, los valor en return deben ser booleanos 
# caso contrario el algoritmo presenta un error el compilar...
# Cuando uno no usa el bool podría retornar cualquier valor, ejemplo
```

     ** la validación de resultado está en que la pila al final debe estar vacia, 
     ya que si la palabra está bien formulada debe tener el mismo número de "(" y ")". 
     Vamos a apilar n veces el valor "(" y si está bien formulado el valor de la entrada s, vamos 
     a tener que desapilar n veces también, por lo tanto la pila va a estar vacia...
     
     ** En casos donde el número de "(" sea mayor al número de ")", entonces en la pila al final 
     va a existir elementos, por lo tanto el algoritmo retornara falso ya que 
     
         len(stack) == 0
         
     ** El otro case sería cuando el número de "(" es menor al número de ")", entonces va a haber un 
     momento donde estemos analizanedo algun ")" y la pila este vacía, osea no existe ningún "(" que le 
     corresponde a ese ")". Al final tenemos una expresión fallida. 
     
     if stack and stack[-1] == "(":           --> len(stack) > 0 and stack[-1] == "(":
           stack.pop() 
     else: 
        return False         
     
     el código if stack es lo mismo que escribit len(stack) > 0: 
     Ahí estamos preguntando si existe algun elemento en stack. Si no existe elemento alguno y estamos 
     analizando ")", entonces la expresión es fallida y retornamos False. 
     






```python
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        
        for char in s: 
            if char in "(":
                stack.append(char)
            elif char == ")":
                if stack and stack[-1] == "(":
                    stack.pop() 
                else: 
                    return False         
                        
        return len(stack) == 0
    
# podemos llamar a esa función en el mismo archivo 

sol = Solution()

print(sol.isValid("(()(((()))))"))
print(sol.isValid("(()(((((())))"))
print(sol.isValid("(()(((()))"))
```
